## __16.1 THE NATURE AND PROPAGATION OF SOUND WAVES__

- A vibrating source in a medium produces sound

- These sound wave are longitudinal in nature and are divided into three categories

    1. ___Infrasonic waves___ - The waves with frequency below the audible range (< 20 Hz)

    2. ___Audible waves___ - The sounds heard by humans. They are between 20 Hz and 20,000 Hz

    3. ___Ultrasonic waves___ - The waves with frequency above the audible range (> 20,000 Hz)

__How Sound Wave Propagates__

- Sound waves propagate in the form of compression and rarefaction of medium

- Due to propagation of these waves, pressure varies periodically in sinusoidal fashion

- A sound wave is explained either by longitudinal displacement suffered by the particles of the medium or by excess pressure generated due to compression and rarefaction

__Wavefront__

- The surface through the points, having same phase of disturbance, it is called wavefront

- For a point source placed in homogenous and isotropic medium, the wavefronts are spherical and the wave is called _spherical wave_

## __16.11 BEATS__

- When two sound waves of same amplitude travelling in same direction with different frequencies superimpose, then intensity varies periodically with time, this effect is called beat

- A complete cycle of maximum intensity and minimum intensity is called one beat and number of beats heard per second is called beat frequency

- Beat frequency is equal to the difference of frequencies of two interfering waves

## __16.12 DIFFRACTION__

- If a plane wave is passed through a small hole, spherical waves are obtained on the other side as if the hole itself is a source sending waves in all directions. Such bending of waves from an obstacle or an opening is called diffraction

## __16.13 DOPPLER EFFECT__

- If a wave source and observer are moving relative to each other, the frequency observed by the observer (f(a)) is different from the actual source frequency (f)

- This phenomenon is known as Doppler's Effect

## __16.14 SOUND BOOM__

- A sonic boom is the sound associated with shock waves created when an object moves through the air at supersonic speeds (faster than Mach 1, ~343 m/s at sea level).

- As the object moves faster than sound, pressure waves cannot spread out normally. Instead, they compress into a cone-shaped shock wave called a Mach cone.

- Observers hear nothing until the shock wave passes over them, at which point the boom is heard all at once, similar to a thunderclap.
