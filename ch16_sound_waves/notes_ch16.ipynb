## __16.1 THE NATURE AND PROPAGATION OF SOUND WAVES__

- A vibrating source in a medium produces sound

- These sound wave are longitudinal in nature and are divided into three categories

    1. ___Infrasonic waves___ - The waves with frequency below the audible range (< 20 Hz)

    2. ___Audible waves___ - The sounds heard by humans. They are between 20 Hz and 20,000 Hz

    3. ___Ultrasonic waves___ - The waves with frequency above the audible range (> 20,000 Hz)

__How Sound Wave Propagates__

- Sound waves propagate in the form of compression and rarefaction of medium

- Due to propagation of these waves, pressure varies periodically in sinusoidal fashion

- A sound wave is explained either by longitudinal displacement suffered by the particles of the medium or by excess pressure generated due to compression and rarefaction

__Wavefront__

- The surface through the points, having same phase of disturbance, it is called wavefront

- For a point source placed in homogenous and isotropic medium, the wavefronts are spherical and the wave is called _spherical wave_
