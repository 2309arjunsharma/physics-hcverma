### __12.1 SIMPLE HARMONIC MOTION__

- Harmonic motion is also known as periodic motion

- When a body repeat its motion continuously on a definite path in a definite interval of time, then its motion is called periodic motion

- The constant interval of time after which the motion is repeated is called the time period of motion (T).

- Eg: 

    - Time period of a minute hand is 60 minutes.

    - Earth takes 365 days to revolve on a definite path around the Sun

    - Earth takes 24 hours to rotate on its axis

- If a body moves to and fro on the same path, it is said to perform oscillations

- Simple Harmonic Motion (SHM) is a special type of oscillation in which the particle oscillates on a straight line, the acceleration of the particle is always directed towards a fixed point on the line and its magnitude is proportional to the displacement of the particle from this point. This fixed point is called _centre of oscillation_

- In oscillatory motion, a force acts on a body to bring it to mean position.

### __12.5 SIMPLE HARMONIC MOTION AS A PROJECTION OF CIRCULAR MOTION__

- Simple Harmonic Motion (SHM) can be visualized as the projection of uniform circular motion onto a straight line. 

- Imagine a particle moving around a circle at constant angular velocity. If you shine a light and observe its shadow on a diameter, the shadow oscillates back and forth in SHM.

- The displacement of this projection varies sinusoidally with time, matching the equations of SHM.

- The restoring force in SHM corresponds to the radial component of circular motion.

- Thus, SHM is essentially the one-dimensional projection of circular motion, linking oscillations of springs, pendulums, and vibrations to rotational dynamics.

- _Thus, the projection of a uniform circular motion on a diameter of the circle is a simple harmonic motion_

### __12.12 DAMPED HARMONIC MOTIONS__

- Damped Harmonic Motion is described as the oscillatory motion of a system where energy is gradually lost due to resistive forces like friction or air resistance.

- Unlike ideal simple harmonic motion, the amplitude of oscillations decreases with time because part of the mechanical energy is dissipated.

- Mathematically, the displacement is expressed as:

    - Œ±: initial amplitude
    - Œ≤: damping constant (rate of decay)
    - œâ: angular frequency of damped oscillation
    - Œ¶: phase constant

- The exponential factor e^{-Œ≤t} shows how amplitude decays over time.

- Depending on the strength of damping, the system may be underdamped (oscillations continue with decreasing amplitude), critically damped (returns to equilibrium fastest without oscillating), or overdamped (returns slowly without oscillations).

- This concept is crucial in real-world systems like car shock absorbers, pendulums in air, and electrical circuits, where damping ensures stability and prevents perpetual oscillations.


If damping is large the system may not oscillate at all. If displaced, it will go towards the mean position and stay there without overshooting on the other side

The damping for which the oscillation just ceases is called _critical damping_

### __12.13 FORCED OSCILLATION AND RESONANCE__

- Forced Oscillation:

    - When a system (like a pendulum or spring-mass) is driven by an external periodic force, it oscillates with the frequency of that force, not its natural frequency. The amplitude depends on the driving frequency and the damping present in the system.
