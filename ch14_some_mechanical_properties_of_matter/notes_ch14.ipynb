## __14.1 MOLECULAR STRUCTURE OF A MATERIAL__

- Matter is made up of molecules and atoms

- An atom is made of nucleus and electrons

- The nucleus contains positively charged protons and nuetrons (nucleons).

- Electromagnetic forces operate between a pair of electrons and between an electron and the nucleus

__Interatomic and Intermolecular Forces__

- A polyatomic molecule is formed when the atoms are arranged in such a fashion that the total potential energy of the system is minimum

- The force between atoms is called interatomic force and the forces between molecules is called intermolecular force

__Bonds__

- The atoms form molecules primarily due to the electrostatic interaction between the electrons and the nuclei.

- These interactions are described in terms of different kinds of bonds

__Ionic Bonds__

- In an ionic bond, two atoms come close to each other and an electron is completely transferred from one atom to the other

- This leaves the first atom positively charged and the other one negatively charged.

- Eg: The positively charged sodium ion and negatively charged chlorine ion attract each other to form an ionic bond resulting in sodium chloride molecule


__Covalent Bond__

- Electrons from neighbouring atoms are made available for sharing between the atoms and is called a covalent bond

- When two hydrogen (H) atoms come close to each other, both the electrons are available to both the nuclei

- Another example is hydrogen chloride (HCl) molecule

__Three States of Matter__

_Solids_

- Definite shape and volume – they don’t change shape unless force is applied.
- Particles are tightly packed – arranged in fixed positions.
- Strong intermolecular forces – particles only vibrate in place.
- Incompressible – cannot be squashed easily.
- High density compared to liquids and gases.

_Liquids_

- Definite volume but no fixed shape – they take the shape of their container.
- Particles are close but can slide past each other – allows flow.
- Moderate intermolecular forces – weaker than solids, stronger than gases.
- Almost incompressible – but slightly more compressible than solids.
- Medium density – usually less than solids but more than gases.


_Gases_

- No definite shape or volume – they expand to fill the container.
- Particles are far apart – move freely and randomly.
- Weak intermolecular forces – negligible compared to solids and liquids.
- Highly compressible – can be squashed easily.
- Low density compared to solids and liquids.

__Solid State__

- The equilibrium positions have a very regular three dimensional arrangement which is called _crystal_

- The positions occupied by the molecules or the ions are called lattice points

__Molecular Solid__

- In a molecular solid, the molecules are formed due to covalent bonds between the atoms

- It depends on whether the molecules are _polar_ or _nonpolar_

- If the centre of negative charge in a molecule coincides with the centre of the poisitive charge, the molecule is _nonpolar_. Eg: Molecules of H, O, Cl, etc

- Otherwise, the molecule is polar. Eg: H₂O

- The bond between polar molecules is called a __dipole-dipole bond__

- The bond between nonpolar molecules is called a __van der Waals bond__

- Molecular solids are usually soft and have low melting point. They are poor conductors of electricity



__Ionic Solid__

- In an ionic solid, the lattice points are occupied by positive and negative ions

- The electrostatic attractions between these ions bind the solid

- These attraction forces are quite strong so that the material is usually hard and has fairly high melting point

- They are poor conductors of electricity

__Covalent solid__

- In a covalent solid the atoms are arranged in the crystalline form

- The neighbouring atoms are bound by shared electrons

- Eg: Diamond, silicon,etc

- They are quite hard, have high melting point and are poor conductors of electricity

## __14.2 ELASTICITY__

- Property of a body by virtue of which it regains its shape and size after removal of deforming force

- If a body completely gains its natural shape after removal of deforming forces, it is called a perfectly elastic body

- If a body remians in the deformed state and does not even partially regain its original shape after removal of the deforming forces, it is called a perfectly inelastic or plastic body

- Steel > Rubber > Plastic

__Microscopic Reason for Elasticity__

- When deforming force is applied on a body, some internal ressistance is developed in body due to intermolecular force, this is called as restoring force

## __14.3 STRESS__

- It is defined as the restoring force acting per unit area.

- Stress (σ) = F(restoring)/Area

- SI = N/m²

- Dimension = [ML-¹T-²]

- Types of stress

    - Longitudinal Stress

    - Shear Stress

    - Volume Stress

## __14.4 STRAIN__

- Strain = Change in the dimension of body/Original dimension of body

- Strain is dimensionless

- Strain is unitless

- Types of strain

    - Longitudinal Strain

    - Shear Strain

    - Volume Strain

## __14.5 HOOKE'S LAW AND THE MODULII OF ELEASTICITY__

- If the deformation is small, the stress in a body is proportional to the corresponding strain

- stress ∝ strain [upto eleastic limit]

- stress = E strain

- E = stress/strain = Modulus of elasticity

Types of modulus

- Young's Modulus

- Shear Modulus

- Bulk Modulus

- Stress Strain Curve

## __14.6 RELATION BETWEEN LONGITUDINAL STRESS AND STRAIN__

__Metal Wire__

- When the strain is small (<0.01), the stress is proportional to the strain

- This is the region where Hooke's Law is valid and where Young's module is defined

- The point on the curve upto which stress and strain are proportional represents the _proportional limit_

- If deformation is increased further, the wire breaks at a point called _fracture point_

- The stress corresponding to this point is called _breaking stress_

- If large deformation takes place between the elastic limit and the fracture point, the material is called _ductile_. If it breaks soon after the elastic limit is crossed, it is called _brittle_

## __14.7 ELASTIC POTENTIAL ENERGY OF A STRAINED BODY__

- When a body is in its natural shape, its potential energy corresponding to the moleculer forces is minimum. When deformed, internal forces appear, and work has to be done against these forces. Thus potential energy of the body is increased. This is called _elastic potential energy_
