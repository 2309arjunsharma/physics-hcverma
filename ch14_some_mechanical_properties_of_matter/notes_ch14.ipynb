## __14.1 MOLECULAR STRUCTURE OF A MATERIAL__

- Matter is made up of molecules and atoms

- An atom is made of nucleus and electrons

- The nucleus contains positively charged protons and nuetrons (nucleons).

- Electromagnetic forces operate between a pair of electrons and between an electron and the nucleus

__Interatomic and Intermolecular Forces__

- A polyatomic molecule is formed when the atoms are arranged in such a fashion that the total potential energy of the system is minimum

- The force between atoms is called interatomic force and the forces between molecules is called intermolecular force

__Bonds__

- The atoms form molecules primarily due to the electrostatic interaction between the electrons and the nuclei.

- These interactions are described in terms of different kinds of bonds

__Ionic Bonds__

- In an ionic bond, two atoms come close to each other and an electron is completely transferred from one atom to the other

- This leaves the first atom positively charged and the other one negatively charged.

- Eg: The positively charged sodium ion and negatively charged chlorine ion attract each other to form an ionic bond resulting in sodium chloride molecule


__Covalent Bond__

- Electrons from neighbouring atoms are made available for sharing between the atoms and is called a covalent bond

- When two hydrogen (H) atoms come close to each other, both the electrons are available to both the nuclei

- Another example is hydrogen chloride (HCl) molecule

__Three States of Matter__

_Solids_

- Definite shape and volume – they don’t change shape unless force is applied.
- Particles are tightly packed – arranged in fixed positions.
- Strong intermolecular forces – particles only vibrate in place.
- Incompressible – cannot be squashed easily.
- High density compared to liquids and gases.

_Liquids_

- Definite volume but no fixed shape – they take the shape of their container.
- Particles are close but can slide past each other – allows flow.
- Moderate intermolecular forces – weaker than solids, stronger than gases.
- Almost incompressible – but slightly more compressible than solids.
- Medium density – usually less than solids but more than gases.


_Gases_

- No definite shape or volume – they expand to fill the container.
- Particles are far apart – move freely and randomly.
- Weak intermolecular forces – negligible compared to solids and liquids.
- Highly compressible – can be squashed easily.
- Low density compared to solids and liquids.
