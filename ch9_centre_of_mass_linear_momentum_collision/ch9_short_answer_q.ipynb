__Q1__

- Yes, COM of a body can be at a point outside the body

__Q2__

- If all the particles lie in the x-y plane, then it is necessary that Centre of Mass also lies in x-y plane

__Q3__

- If all the particles of a system lie in a cube then it is necessary that the COM be in a cube
- For 2 particle system COM always lies in between them on the line joining the particles following the same logic we can say that COM lie inside the cube
