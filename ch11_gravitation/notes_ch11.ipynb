### 11.1 HISTORICAL INTRODUCTION

- Famous Indian Astronomer, ___Aryabhatt___, studied the motions of planets, Sun, moon, etc, in 5th century A.D., and wrote his conclusions in his book ___Aryabhatiya___

- The hypothesis about planetery motion was given by ___Nicolaus Copernicus___ (1473-1543)

- The careful experimental measurements of the positions of the planets and the Sun by ___Tycho Brahe___ (1546-1601)

- Analysis of the data and the formulation of empirical laws by ___Johannes Kepler___ (1571-1630).

- In the mid-1660s, Newton reflected on why apples fall straight down. He realized the same force pulling the apple to Earth also governs celestial bodies. By 1687, in his ___Principia Mathematica___, Newton formulated the law of universal gravitation: every mass attracts every other mass, with force proportional to their masses and inversely proportional to the square of their distance. This unified terrestrial and celestial mechanics, explaining both falling objects and the Moon’s orbit. Newton’s insight transformed science, replacing earlier ideas from Aristotle and Galileo with a precise mathematical framework. His theory remained dominant until Einstein’s relativity refined gravity’s nature centuries later.

- So, the development of the general theory was done by ___Isaac Newton___ (1642-1727)

__Laws of Kepler__

1. All planets move in elliptical orbits with the sun at the focus

2. The radius vector from the sun to the planet sweeps equal area in equal time

3. The square of a time period of a planet is proportional to the cube of the semimajor axis of the ellipse.

### 11.2 MEASUREMENT OF GRAVITATIONAL CONSTANT G



__What is G?__

- G is numerically equal to the force of attraction between two unit masses placed at unit distance apart

-  F = G(m₁m₂/r²)

- If m₁ = m₂ = 1 and r = 1, then F = G

- G = 6.67 X 10-¹¹ N-m²/kg²

- Dimension = [M-¹L³T-²]
