### 11.12 WEIGHTLESSNESS IN A SATELLITE

- Astronauts feel weightless in a satellite because both they and the satellite are in continuous free fall toward Earth, but since they move forward fast enough, they keep missing Earth. This creates the sensation of weightlessness, even though gravity is still acting on them.

- Why Weightlessness Happens in Satellites
    - Gravity is still present: Satellites orbit Earth because gravity pulls them inward. Without gravity, they would fly off into space.
    - Free fall condition: A satellite and everything inside it (including astronauts) are falling toward Earth at the same rate.
    - Forward motion: The satelliteâ€™s high speed ensures it keeps missing Earth, creating a stable orbit.
    - No support force: On Earth, you feel weight because the ground pushes up on you. In orbit, thereâ€™s no surface pushing back, so astronauts feel weightless.

- Common Misconceptions
