## __15.1 WAVE MOTION__

- _Wave motion_ is the transfer of energy through a medium without permanent displacement of particles. Particles oscillate around fixed positions, creating disturbances that travel as mechanical or electromagnetic waves. Examples include sound, water waves, and light, all showing rhythmic propagation of energy.

- The transfer of energy and momentum without the actual displacement of the medium particle is called wave

- Types of waves

    1. Based on requirement of medium

    2. Based on motion of the medium particles

    3. Based on energy propagation

    4. Based on direction of propagation

__Based on requirement of medium__

- __Mechanical waves__ - These waves requiring a medium. Eg: Sound waves,waves on a string

- __Electromagnetic or nonmechanical__ - those which do not require a medium. Eg: Radio waves, Light waves

__Based on motion of the medium particles__

- __Longitudinal waves__ - Particles of the medium vibrate in the direction of wave motion. Eg: Sound waves, waves on a spring

- __Transverse waves__ - Particles of the medium vibrate perpendicular to the direction of wave motion. Eg: Light waves, waves on a string


__Based on energy propagation__

- __Progressive waves__ - Waves propagate with fixed velocity in a medium. Eg: Waves on liquid surface

- __Stationary waves__ - The particle of the medium vibrate with different amplitude but energy does not propagate. Eg: Vibration of sonometer wire

__Based on direction of propagation__

- __One dimesional waves__ - Single direction propagation. Eg: Stretched string

- __Two dimensional waves__ - Wave propagation in a plane. Eg: Waves on liquid surface

- __Three dimensional waves__ - Waves in all directions. Eg: Sound waves

## __15.2 WAVE PULSE ON A STRING__

- When a disturbance is localised only to a small part of the space at a time, we say that a _wave pulse_ is passing through that part of the space

- If the source is active for some extended time repeating its motion several times, we get a _wave train_ or a _wave packet_

__Equation of a Travelling Wave__

- It is a differential equation whose solution is given by:-

- y = f(ax +- bt)

- Where a and b are constant

## __15.3 SINE WAVE TRAVELING ON A WAVE__

- The segment where disturbance is positive is called a _crest_ of the wave

- The segment where disturbance is negative is called a _trough_ of the wave

- The distance between consecutive crests and troughs is called _wavelength_(λ)

- λ = 2πν/ω = νT

- The quantity 2π/λ is called the _wavenumber_ (k)

## __15.8 REFLECTION AND TRANSMISSION OF WAVES__

- If a wave enters a region where the wave velocity is smaller, the reflected wave is inverted

- If it enters a region where the wave velocity is larger, the reflected wave is not inverted

- The transmitted wave is never inverted

## __15.11 ANALYTIC TREATMENT OF VIBRATION OF A STRING FIXED AT BOTH ENDS__

- An integral multiple of a frequency is called its _harmonic_.

- Thus, _for a string fixed at both the ends, all the overtones are harmonics and all the harmonics of the fundamental frequency are overturns_

- This property is unique to the string and makes it so valuable in musical instruments such as violin, guitar, sitar, santoor, sarod

## __15.13 LAWS OF TRANSVERSE VIBRATIONS OF A STRING: SONOMETER__

- "Laws of transverse vibrations of a string"

    1. __Law of length__ - The fundamental frequency of vibration of a string (fixed at both ends) is inversely proportional to the length of the string provided its tension and its mass per unit length remain same

    2. __Law of tension__ - The fundamental frequency of a string is proportional to the square root of its tension provided its length and the mass per unit length remain the same

    3. __Law of mass__ - The fundamental frequency of a string is inversely proportional to the square root of the linear mass density, i.e., mass per unit length, provided the length and tension remain the same

- These laws may be experimentally studied with an apparatus called _sonometer_
