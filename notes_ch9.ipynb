# __CH-9 CENTRE OF MASS, LINEAR MOMENTUM, COLLISION__


## 9.1 CENTRE OF MASS

- Centre of mass of two particles lies on the line joining them
    - The centre of mass divides the distance between two particles in the inverse ratio of masses
- r₁/r₂ = m₂/m₁
- r(cm) = (m₁r₁ + m₂r₂)/(m₁ + m₂)
- For ring and semi-circular ring, Centre of Mass lies outside the body.

__Centre of Mass of Composite Bodies__
- If surface mass density (mass per unit area) is denoted by a then masses of bodies assumed to be uniform.
- (1-D) Linear mass density = m/L
- (2-D) Surface mass density = m/A
- (3-D) Volume mass density = m/V
