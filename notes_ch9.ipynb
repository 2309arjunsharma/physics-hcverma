## 9.1 CENTRE OF MASS

- Centre of mass of two particles lies on the line joining them
    - The centre of mass divides the distance between two particles in the inverse ratio of masses
- r₁/r₂ = m₂/m₁
- r(cm) = (m₁r₁ + m₂r₂)/(m₁ + m₂)
- For ring and semi-circular ring, Centre of Mass lies outside the body.

__Centre of Mass of Composite Bodies__
- If surface mass density (mass per unit area) is denoted by a then masses of bodies assumed to be uniform.
- (1-D) Linear mass density = m/L
- (2-D) Surface mass density = m/A
- (3-D) Volume mass density = m/V

### 6.2 COLLISION
- Collision is a kind of interaction between two or more bodies which come in contact with each other for a very short time interval

### 6.3 COEFFICIENT OF RESTITUTION

- The ratio of magnitudes of impulse of restitution to that of deformation is called the coefficient of restitution
-  e = impulse of recovery/impulse of deformation
- e = velocity of separation/velocity of approach = [ V(B) - V(A) ]/[ U(A) - U(B) ]

### 6.4 TYPES OF COLLISION
- __Collision__
    - __On the basis of impact__
        - __Head on collision__
        - __Oblique collision__
    
    - __On the basis of energy__
        - __Elastic collision__
            - After collision, bodies regain their shape and size (PE = 0)
            - Energy is conserved
            - e = 1
            - Momentum conservation
            - KE(I) = KE(f)
            - For elastic collision, if masses are same then, the particles will exchange their velocitites
        - __Inelastic collision__
            - m₁u₁ + m₂u₂ = m₁v₁ + m₂v₂
            - Energy conservation is not valid here as the shape of the bodies changes
            - There is loss of KE
            - 0 < e < 1
        - __Perfectly Inelastic collision__
            - e = 0
