## __13.1 FLUIDS__

- Liquids and gases together are called fluids, i.e., those which can flow

- These liquids are mainly incompressible and nonviscous.

- Conditions

    - The first condition mans that the density of the liquid is independent of the variations in pressure and always remains constant

    - The second condition means that parts of the liquid in contact do not exert any tangential force on each other.
    

__Relative Density__

- It is defined as the ratio of density of given fluid to the density of the pure water at 4°C.

- Relative Density = Density of given fluid/ Density of pure water at 4°C.

- It is unitless

- Dimensionless

- Density of water at 4°C = 1000 kg/m³ = 1 gm/cm³
