## __13.1 FLUIDS__

- Liquids and gases together are called fluids, i.e., those which can flow

- These liquids are mainly incompressible and nonviscous.

- Conditions

    - The first condition mans that the density of the liquid is independent of the variations in pressure and always remains constant

    - The second condition means that parts of the liquid in contact do not exert any tangential force on each other.
    

__Relative Density__

- It is defined as the ratio of density of given fluid to the density of the pure water at 4°C.

- Relative Density = Density of given fluid/ Density of pure water at 4°C.

- It is unitless

- Dimensionless

- Density of water at 4°C = 1000 kg/m³ = 1 gm/cm³

__Specific Gravity__

- It is defined as the ratio of specific weight of the given fluid to the specific weight of pure water at 4°C

- Specific Gravity = Specific weight of given fluid/Specific weight of pure water at 4°C

## __13.2 PRESSURE IN A FLUID__

P = F/ΔS

1 atm = 1.01 X 10⁵ Pa
1 bar = 10⁵ Pa

__Unit of Pressure__

The SI unit of pressure in Nm-² called Pascal (Pa)

__Variation of Pressure:__

- 1. In fluids at rest or at uniform speed

    - In horizontal plane

    - With height or depth

- 2. Variation of pressure at two points in different height/depth

- 3. In fluids under acceleration

    - In horizontal direction

    - In vertical direction

    - In rotation

## __13.3 PASCAL'S LAW__

- Pascal's Law is stated by:

    - A liquid exerts equal pressure in all directions.

    - If the pressure in an enclosed fluid is changed at a particular point, the change is transmitted to every point of the fluid and to the walls of th container without being diminished in magnitude
    

__Applications of Pascal's Law__

- Hydraulic jacks, Hydraulic lifts, Hydraulic press, hydraulic brakes, etc

- Hydraulic Lift:

    - Pressure transmitted = F₁/A₁
    
    - Pressure Transmitted = F₂/A₂

- Since Pressure is equally transmitted,

- Therefore, 
    - F₁/A₁ = F₂/A₂

## __13.4 ATMOSPHERIC PRESSURE AND BAROMETER__

- Atmospheric pressure is the force exerted by the weight of air molecules pressing down on the Earth’s surface.

- It is measured at sea level as about 101,325 pascals (Pa), or 1 atmosphere (atm). 

- It decreases with altitude because there is less air mass above

__Barometer__

- Torricelli deviced this instrument to measure atmospheric pressure.

- In this, a glass tube is completely filled with mercury and dipped into a mercury filled beaker with its open and inside mercury

- The upper part contains vacuum as the mercury goes down and no air is allowed in.

__Manometer__

- Manometer is a simple device to measure gauge pressure of fluids

- It consists of a U tube having some liquid. One end of the tube is open to the atmosphere and the other end is connected to a vessel(of which we have to measure the gauge pressure)

## __13.5 ARCHIMEDES' PRINCIPLE__

__Buoyancy__

- If a body is partially or fully immersed in a fluid, it experiences an upward force due to fluid surrounding it.

- The phenomenon of force exerted by fluid on the body is called buoyancy and the force is called buoyant force

- When a solid is immersed in a liquid, it displaces the liquid, partially or completely of its own volume. While doing so it pushes or applies a downward force on the liquid, hence, in accordance to Newton's third law, the liquid also exerts an upward force(equal to the weight of the displaced liquid) 

- Archimedes' principle states that when a body is partially or fully dipped  into a fluid at rest, the fluid exerts an upward force of buoyancy equal to the weight of the displaced fluid

- If V be the volume of displaced fluid of density ρ then,

    - buoyant force = Vρg = F(B)

__Floatation__

- When a solid body is dipped into a fluid, the fluid exerts an upward force of buoyancy on the solid. If the force of buoyancy equals the weight of the solid, the solid will remain in equilibrium. This is the law of floatation

## __13.8 STEADY AND TURBULENT FLOW__

- A flow is said to be steady if the velocity, pressure and density at any point in the flow do not change with time.

- In an unsteady flow, the velocity at a point in the flow varies with time 

- Steady flow is also called streamline flow

__Line of Flow: Streamline__

- The path taken by a particle in flowing fluid is called its line of flow

__Tube of Flow__

- When a liquid is passed slowly trhough a pipe, the pipe itslef is one tube of flow

## __13.11 BERNOULLI EQUATION__

- According to Bernoulli's theorem, the sum total of these three kinds of energies kinetic energy, potential energy and pressure energy remains constant along a streamline in a steady flow of an ideal fluid.

- It is a mathematical expression of the law of conservation of mechanical energy in fluid dynamics.

- Total energy per unit volume(A) = Total energy per unit volume(B)

- Bernoulli equation relates the speed of a fluid at a point, the pressure at that point and the height of that point above a reference level.

- It is just the application of work-energy theorem in the case of fluid flow.

## __13.12 APPLICATIONS OF BERNOULLI EQUATION__

__(a) Hydrostatics__

- If the speed of the fluid is zero everywhere, we get the situation of hydrostatics

__(b) Speed of Efflux__

- The speed of liquid coming out through a hole at a depth h below the free surface is the same as that of a particle fallen freely through the height h under gravity

- This is known as Torricelli's theorem.

- The speed of the liquid coming out is called the speed of efflux
