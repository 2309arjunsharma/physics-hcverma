## 8.1 KINETIC ENERGY

- Quantitatively the energy of the moving particle (over and above its energy at rest) is called kinetic energy
- A force is unnecessary to change the kinetic energy of a particle
- Kinetic energy is the internal capacity of doing work of an object by virtue of its motion

## 8.2 WORK AND WORK-ENERGY THEOREM

- __Work__= Whenever a force acting on a body displaces it, work is said to be done by the force


__Work done by variable force__
- If the force applied on a body is changing in direction or magnitude or both, the force is said to be variable.
- If dr = small displacement
- F = variable force acting on body

__Work energy theorem__
- Work done by all the forces (conservative, non conservative, external or internal) acting on a particle or an object is equal to the change in its kinetic energy
- work done by all the forces = change in KE
- W = ΔKE = 1/2mv²(f) - 1/2mv²(i)

## 8.3 CALCULATION OF WORK DONE

__Positive work__
- If angle 0 is acute (0 < 90) then work is said to be positive 
- Positive work signifies that external force favors the motion of the body

__Examples__
- When the body falls freely, the work done by gravity is positive
- When spring is stretched, stretching force and displacement both are in same direction. So work done by stretching force is positive

## 8.4 POTENTIAL ENERGY
- The energy possessed by a body by virtue of its position or configuration in a conservation force field
- Potential energy is a relative quantity
- Potential energy of a body in a conservative force field is equal to the work done by conservative force in moving the body from its present position to reference position
- At a certain reference position, the potential energy of the body is assumed to be zero.


## 8.5 CONSERVATIVE AND NON CONSERVATIVE FORCES

- If the work done by a force during a round trip of a system is always zero, the force is said to be conservative. Otherwise, it is called non-conservative
- Conservative eg: Coulomb force and force of spring
- Non-conservative eg: force of friction

## 8.6 DEFINITION OF POTENTIAL ENERGY AND CONSERVATION OF MECHANICAL ENERGY

1. Work done on a particle is equal to the change in its kinetic energy

2. Work done on a system by all the (external and internal) forced is equal to the change in its kinetic energy

3. A force is called conservative if the work done by it during a round trip of a system is always zero. The force of gravitation, Coulomb force, force by a spring, etc are conservative.
If the work done by it during a round trip is not zero, the force is nonconservative. The force of friction is nonconservative.

4. The change in the potential energy of a system corresponding to conservative internal forces is equal to negative of the work done by these forces.

5. If no external forces act (or work done is zero) and the internal forces are conservative, the mechanical energy of the system remains constant. This is known as principal of conservation of mechanical energy.

6. If some of the internal forces are nonconservative, the mechanical energy of the system is not constant.

7. If the internal forces are conservative, the work done by the external forces is equal to the change in mechanical energy.

## 8.7 CHANGE IN THE POTENTIAL ENERGY IN A RIGID-BODY-MOTION
- The potential energy of a system depends only on the separation between the interacting particles

## 8.8 GRAVITATIONAL POTENTIAL ENERGY
