### 10.1 ROTATION OF A RIGID BODY ABOUT A GIVEN FIXED LINE

- A body for which the separation between the particles remains constant is a rigid body

- Motion of a body in such a way that the particles of the body performs circular motion about a common centre
- The time period and angular velocity is same for the particles

- __Types of motion__
    - Translational motion
        - Particle follows linear motion
    - Rotational motion
        - Particle follows circular motion
    - Translational + Rotational motion
        - It is also called Rolling motion
        - Particle follows linear and circular motion

- __Axis of Rotation__
    - An imaginary line perpendicular to the plane of circular paths of particle of a rigid body in rotation and containing the centres of all these circular path is known as axis of rotation

- Examples of objects which rotate about a fixed axis
    - Fan
    - Wheel (Also includes translation)

### 10.8 ANGULAR MOMENTUM

- Angular Momentum [L]
- L = Iω
- It is the Moment (Rotation) of Momentum
- L = r X p = r X mv = m(r X v)
- L = mvr sinθ where θ is the angle between v and r




- Angular momentum of a body about a given axis is the product of its linear momentum and perpendicular distance of line of action of linear momentum vector from the axis of rotation

### 10.9 L = _I_ ω

- It is the formula of angular momentum

- Angular momentum is an axial vector i.e. always directed perpendicular to the plane of rotation and along the axis of rotation
- SI unit = kg-m²-s-¹ or J-sec
- Dimension = [ML²T-¹]

### 10.10 CONSERVATION OF ANGULAR MOMENTUM

- Radial component of momentum (p) = p cosθ

- Transverse component of momentum (p) = p sinθ

- We know that L = Iω

- Differentiating w.r.t time
- dL/dt = τ

- Tis is Newton's second law in rotation

- Angular momentum of a system (May be particle or body) remains constant if resultant torque acting on it is zero
- So if τ = 0, then L = Iω is constant
- ∴ I ∝ 1/ω

### 10.14 CALCULATION OF MOMENT OF INERTIA

- The moment of inertia is the rotational equivalent of mass in linear motion.
- It measures how difficult it is to rotate an object about an axis. Just like heavier objects resist changes in motion, objects with larger moment of inertia resist changes in rotation.
- It depends on two factors: the mass of the body and how that mass is distributed relative to the axis of rotation.
- Mathematically, I=\sum mr^2, where m is mass and r is distance from the axis. For example, a rod rotated about its end has greater inertia than when rotated about its center.
