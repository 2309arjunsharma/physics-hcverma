# __CH-10 ROTATIONAL MECHANICS__

### 10.1 ROTATION OF A RIGID BODY ABOUT A GIVEN FIXED LINE

- A body for which the separation between the particles remains constant is a rigid body

- Motion of a body in such a way that the particles of the body performs circular motion about a common centre
- The time period and angular velocity is same for the particles

- __Types of motion__
    - Translational motion
        - Particle follows linear motion
    - Rotational motion
        - Particle follows circular motion
    - Translational + Rotational motion
        - It is also called Rolling motion
        - Particle follows linear and circular motion

- __Axis of Rotation__
    - An imaginary line perpendicular to the plane of circular paths of particle of a rigid body in rotation and containing the centres of all these circular path is known as axis of rotation

- Examples of objects which rotate about a fixed axis
    - Fan
    - Wheel (Also includes translation)

### 10.2 KINEMATICS

- v = u + at
- s = ut + 1/2 at²
- v² = u² + 2as
- s(n)th = u + a/2(2n - 1)
- s = (u + v)t/2
- ω = ω₀ + αt
- θ = ω₀t + 1/2αt²
- ω² = ω₀² + 2αθ 
- θ(n)th = ω₀ + a/2(2n - 1)
- θ = (ω + ω₀)t/2


__SYMBOLS__
- Linear (L), Angular (A)
- Speed
    - L = u , A = ω₀
    - L = v, A = ω
- Displacement
    - L = S, d , A = θ
- Acceleration
    - L = a, A = α
- Time
    - L = t, A = t
