## __Ch-1 INTRODUCTION TO PHYSICS__

## 1.1 What is Physics?

- Physics is the study of nature and its laws
- Revealing the laws of nature from observed events is Physics
- Physics is concerned with the basic rules which are applicable to all domains of life
- Understanding of physics, therefore, leads to applications in many fields including bio and medical sciences
- Physics is like big chess game played by Nature.
- We have only discovered the rules through famous physicists like Aryabhat, Newton, Einstein or Feynman
- They are great physicists because from the observations available at that time, they could guess and frame the laws of physics which explained these observations

## 1.2 Physics and Mathematics

- The description of nature becomes easy if we have freedom to use mathematics.
- Gravitational Force = m₁m₂/r²
- Mathematics is the language of physics.
- Algebra, trignometry, calculas, etc are used

## 1.3 Units

- Physics describes laws of nature.
- This is quantitative and involves measurement and comparision of physical quantities.
- We use an unit mass
- The physical quantity are quantitavely expressed in terms of a unit of that quantity.

__Who decides Units?__
- A body named ___Conférence Générale des Poids et Measures___ or ___CGPM___, also known as ___General Conference on Weight and Measures___.

__Fundamental and Derived Quantities__
- We can define a set of fundamental quantities as follows-

    (a) the fundamental quantities should be independent of each other, and
    
    (b) all other quantities may be expressed in terms of the fundamental quantities.
- It turns out that the number of fundamental quantities is only seven.
- One can take length (distance) and time as fundamental quantities, then speed is a derived quantity (derived from fundamental quantities by multiplication and division).
- The units defined for the fundamental quantities are called fundamental units and those obtained for the derived quantities are called derived units.
- Fundamental quantities are also called base quantities.

__SI Units__
- In 1971, CGPM held its meeting and decided a system of units, known as the ___International System of Units or SI___. Its French Name is ___Le Systéme International d'Unités___.

__Quantity -- Name of Unit -- Symbol__

- Besides the seven fundamental units, two supplementary units are defined. They are for plane angle and solid angle. The unit for plane angle is ___radian___ with the symbol ___rad___ and the unit for the solid angle is ___steradian___ with the symbol ___sr___.

## 1.4 Definitions of base units

- Any standard unit should have -

    (a)___Invariability___: The standard unit must be invariable. Thus, defining distance between the tip of the middle finger and the elbow as a unit of length is not invariable.

    (b)___Availabilty___: The standard unit should be easily made available for comparing with other quantities.
- CGPM decided in 2018 that all the SI base quantities will be defined on the basis of certain universal constants. In this case both the criteria of invariablity and availability are automatically satisfied.
