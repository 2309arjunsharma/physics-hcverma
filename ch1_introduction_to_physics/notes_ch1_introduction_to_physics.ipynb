## 1.3 Units

- Physics describes laws of nature.
- This is quantitative and involves measurement and comparision of physical quantities.
- We use an unit mass
- The physical quantity are quantitavely expressed in terms of a unit of that quantity.

__Who decides Units?__
- A body named ___Conférence Générale des Poids et Measures___ or ___CGPM___, also known as ___General Conference on Weight and Measures___.

__Fundamental and Derived Quantities__
- We can define a set of fundamental quantities as follows-

    (a) the fundamental quantities should be independent of each other, and
    
    (b) all other quantities may be expressed in terms of the fundamental quantities.
- It turns out that the number of fundamental quantities is only seven.
- One can take length (distance) and time as fundamental quantities, then speed is a derived quantity (derived from fundamental quantities by multiplication and division).
- The units defined for the fundamental quantities are called fundamental units and those obtained for the derived quantities are called derived units.
- Fundamental quantities are also called base quantities.

__SI Units__

## 1.4 Definitions of base units

- Any standard unit should have -

    (a)___Invariability___: The standard unit must be invariable. Thus, defining distance between the tip of the middle finger and the elbow as a unit of length is not invariable.

    (b)___Availabilty___: The standard unit should be easily made available for comparing with other quantities.
- CGPM decided in 2018 that all the SI base quantities will be defined on the basis of certain universal constants. In this case both the criteria of invariablity and availability are automatically satisfied.

__Second__
- 1 second is the time that makes the unperturbed ground state hyperfine transition frequency to be △vCs 9192631770 when expressed in the unit Hz which is equal to s-¹.
- A second is defined as the time it takes for a cesium-133 atom to undergo 9,192,631,770 energy transitions in a specific state.
- A second is the time it takes for a tiny atom called cesium to vibrate 9,192,631,770 times. Scientists use this to measure time very accurately!

__Metre__
- 1 metre is the length that makes the speed of light in vacuum to be 299792458 when expressed in the unit m.s-¹, where the second is defined in terms of the caesium frequency △vCs.
- A meter is the distance light travels in 1/299,792,458 seconds in empty space. Scientists use this to measure length very precisely!
 
__Kilogram__
- 1 kilogram is the mass that makes the Planck's constant h to be 6.62607015 X 10-³⁴ when expressed in the unit J.s which is equal to kg.m²s-¹, where the metre and the second are defined in terms of c and △vCs.
- A kilogram is the amount of matter that makes a special number, called Planck’s constant, exactly 6.62607015 × 10⁻³⁴ in scientific calculations. Scientists use this to measure weight very precisely!

__Ampere__
- 1 ampere is the current which makes the elementary charge e to be 1.602176634 X 10-¹⁹ when expressed in the unit C which is equal to A.s, where the second is defined in terms of △vCs.
- The ampere (A) measures how much electricity is flowing. One ampere means a tiny charge (like a super small droplet of electricity) moves at a fixed rate, and scientists have given it a special number (1.602176634 × 10⁻¹⁹) to keep everything accurate.

__Kelvin__
- 1 kelvin is the temperature that makes the Boltzmann constant to be 1.380649 X 10-²³ when expressed in the unit J.K-¹ which is equal to kg.m²s².K-¹, where kilogram, metre, and second are defined in terms of h,c and △vCs.
- Kelvin (K) is a way to measure temperature, just like Celsius or Fahrenheit. But instead of talking about "hot" or "cold," scientists use kelvin to describe temperature in terms of energy. The Boltzmann constant (k) is a special number that connects temperature and energy. When we say 1 kelvin, it means that energy follows a fixed rule, and this rule is based on the Boltzmann constant (1.380649 × 10⁻²³).

__Mole__
- 1 mole of a substance is defined to contain exactly 6.02214076 X 10²³ elementary entities. This number is the fixed numerical value of the Avogadro constant Na, when expressed in the unit mol-¹ and is called Avogadro number.
- A mole (mol) is a way to count really tiny particles, like atoms or molecules, in chemistry. Instead of counting each one individually (which would take forever!), scientists use a special number called the Avogadro number: 6.02214076 × 10²³
- This means 1 mole of a substance always has exactly this many particles, no matter what the substance is—whether it's water, oxygen, or sugar!

__Candela__
- The candela is the SI unit of luminous intensity. 1 candela is the luminous intensity that makes the luminous efficacy of monochromatic radiation of frequency 540 X 10¹² Hz, Kcd to be 683 when expressed in the unit lm.W-¹ which is equal to cd.sr.kg-¹m²s³, where kilogram, metre and second are defined in terms of h,c and △vCs.
- The candela (cd) is a unit that measures how bright a light is. Scientists use it to describe how much light comes from a source in a specific direction.
To keep things consistent, they define 1 candela based on a special type of light with a fixed frequency (540 × 10¹² Hz) and a luminous efficacy (how efficiently the light produces brightness) of 683 lumens per watt.
- 1 candela is a unit of brightness that helps scientists measure how much light a source gives off! 


## 1.5 Dimensions

- The exponent of a base quantity that enters a expression, is called the ___dimension of the quantity in that base___.
- Force is equal to mass times accerleration. 
- Acceleration is change in velocity divided by time interval. 
- Velocity is length divided by time interval.
- F = ma, M X V/T, M X (L/T)/T, F = MLT-²
- Thus dimensions of force are 1 in mass, 1 in length, -2 in time. So, dimensions are (1,1,-2)
- Mass is denoted by M, length by L, time by T, current by I. Temperature by K, luminous intensity by cd, amount of substance by mol.
- Dimensional fromula = [M¹L¹T¹] where the number 1 can be replaced with any power.
- Pa (Pressure)=N/m2 
- = MLT-²/L²
- = M¹L¹-²T-²
- = M1L-1T-2 (Pressure Dimensional Formula)
- Dimension= 1,-1,-2

## 1.6 Uses of Dimensions
__A. Homogeneity of Dimensions in an equation__
- If in a given relation, the terms on both the sides have the same relation, then the relation is dimensionally correct. This is known as the ___"principle of homogeneity of dimensions"___.
- Eg: To check whether v = u + at is correct or not--
- dimension of -- LHS = RHS then the given equation is correct
- LT-¹= LT-¹ + LT-²T
- LT-¹ = LT-¹ + LT-¹
- This equation is dimensionally correct

__Limitations of the Homogeneity method__
- Eg: S = ut + at² (Incorrect formula)
- length[L] = LT-¹T¹ + LT-²T²
- [L] = [L] + [L]
- Since, LHS = RHS
- Yes, this equation is dimensionally correct but numerically incorrect.
- Real formula- S = ut + 1/2at²
- This equation is dimensionally and numerically correct.
- 1/2 is dimensionless (Dimension = 0)

__B. Conversion of Units__
- This is based on a fact that magnitude of a physical quantity remains same whatever system is used for measurement.
- ___Magnitude = numeric value(n) X unit(u) = constant(K)___
- Or ___n₁u₁=n₂u₂___ 
- ___Eg: Area=5m² ,Convert to cgs___
- n₁u₁=n₂u₂
- n₂=n₁ X u₁/u₂
- area=side²=L²
- n₂=5 m/cm=5[L₁/L₂]²=5[m/cm]²
- 5 [100cm/cm]² = 5 X 10000 = 50000 cm² = 5 X 10⁴ cm²

__C. Deducing Relation among the Physical Quantities__
- Sometimes dimensions can be used to deduce relation between the physical quantities.
- If one knows the quantities on which a particular physical quantity depends and if one is given that this dependence is of product type. Then method of dimension may be helpful to derive relation
