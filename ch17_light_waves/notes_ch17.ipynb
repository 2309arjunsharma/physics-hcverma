## __17.1 WAVES OR PARTICLES__

__Newton__

- Newton believed that light is a collection of particles.

- He believed that a light source emits tiny corpulses of light and these corpuscles travel in straight lines when not acted upon by external forces

- He had performed a number of experiments to study the behavior of light

- The book _Optiks_ written by him, gives a classic account of these experiments

__Huygens__

- The Dutch physicist Christian Huygens (1629-1695) suggested that light may be a wave phenomenon

- The apparent rectilinear propagation of light may be due to the fact that the wavelength of light may be much smaller than the dimensions of these openings and obstacles

__Young__

- Thomas Young (1773-1829) performed a double slit experiment in 1801

- This experiment clearly established that light coming from two coherent sources interferes and produces maxima and minima depending on the path difference



__Fresnel__

- A series of experiments on diffraction of light conducted by the French physicist Augustin Jean Fresnel (1788-1827)

__Foucault__

- Measurement of velocity of light in water by Foucault in 1850

__Maxwell__

- Development of theory of electromagnetic waves by Maxwell in 1860 which correctly predicted the speed of light, were parts of a long activity which put the corpuscle theory of light to an end and convincingly established that light is a wave phenomenon

__Hallwachs and Lenard__

- The wave theory of light failed to explain Hallwachs and Lenard's observation in 1900 that when light falls on a metal surface, electrons are ejected and that the kinetic energy of the emitted electrons does not depend on the intensity of the light used

- __Hertz__ possibly had the first observation of this phenomenon in early 1880's. This observation is known as _photoelectric effect_

__Einstein__

- Photoelectric effect was explained by Albert Einstein in 1905 on a particle model of light only.

## __17.2 THE NATURE OF LIGHT WAVES__

- The light waves need no material medium to travel 

- They can propagate in vacuum

- Light is a nonmechanical wave

- When a light wave travels in a transparent material, the speed is decreased by a factor μ, called the _refractive index_ of the material

- μ = speed of light in vacuum/speed of light in material

- Light of single wavelength is called _monochromatic light_

- _Geometrical optics_ studies light as rays, explaining reflection, refraction, image formation, and lens behavior using geometry, without considering wave or quantum effects, simplifying analysis of mirrors and lenses.

1. ___Rectilinear Propagation of Light___: Light travels in straight lines unless it is reflected by a polished surface or the medium of propagation is changed

2. ___Reflection of light___: 
    1. The angle of incidence and reflection (i.e., the angles made by the incident and reflected rays with the normal to the surface) are equal.
    
    2.  Also, the incident ray, the reflected ray and the normal to the reflecting surface are coplanar

3. ___Refraction of light___: When light travelling in one medium enters another medium, the nagle of incidence _i_ and the angle of reflection _r_ (angle made by the refracted ray and normal) satisfy 

- sin _i_/sin _r_ = v₁/v₂

- where v₁ and v₂ are the speeds of light in different mediums.

- Also, the incident ray, the refracted ray and the normal to the surface are coplanar

## __17.7 INTERFERENCE FROM THIN FILMS__

- A thin film is a very thin layer of a substance whose thickness is comparable to the wavelength of light (about 10-⁷ to 10-⁶ m)

- Examples:
    
    - Soap bubbles
    - Oil layer on water
    - Anti-reflection coating on glasses

- When light falls on a thin film:

    1. Some light reflects from the top surface
    2. Some light enters the film, reflects from the bottom surface, and comes out

- These two reflected rays combine → this is called interference

- The two rays travel different paths

- Sometimes they add up (_constructive interference_) → bright color

- Sometimes they cancel (_destructive interference_) → dark region

- Different wavelengths (colors) interfere differently → rainbow pattern

- When light reflects from a denser medium, it gets a phase change of π (180°)

- This plays a crucial role in deciding bright/dark patterns

## __17.8 FRESNEL BIPRISM__

- A Fresnel biprism is an optical device made by joining two very thin prisms base-to-base.

- It is used to produce interference fringes by creating two virtual coherent sources from a single slit source.

1. A monochromatic light source passes through a narrow slit (S)

2. Light falls on the biprism

3. The biprism bends (refracts) light in two directions

4. This creates two virtual sources 𝑆₁ and 𝑆₂

- These sources are:

    - Coherent (same frequency & constant phase difference)

    - Very close to each other

5. Light from 𝑆₁ and 𝑆₂ overlaps on the screen → interference pattern (fringes)

- β = λD/d
	​
- Where:

    - β = fringe width

    - λ = wavelength of light

    - D = distance between source and screen

    - d = distance between virtual sources 𝑆₁ and 𝑆

## __17.9 COHERENT AND INCOHERENT SOURCES__

- Two sources of light waves are said to be _coherent_ if the initial phase difference δ₀ between the waves emitted by the sources remains constant in time

- If δ₀ changes randomly with time , the sources are called _incoherent_
